## Course Description

In this class, we will focus on solving numerical linear algebra problem related to real world applications. We will learn to use Python and Jupyter Notebooks in a **top-down** approach. That means, we will start using some "black boxes" or matrix decompositions that haven't yet been explained, and then we'll dig into the lower level details later. The following topics may be covered through this course:
* Modeling with NMF and SVD
* Background Removal with Robust PCA
* Compressed Sensing for CT scans with Robust Regression
* Predicting Health Outcomes with Linear Regression
* PageRank with Eigen Decompositions
* QR Factorization
* Iterative Methods for Sparse Systems

## Student Learning Outcomes

Upon the completion of the course, you will be able to 
1. install various python modules and look up useful functions through documentations;
2. use python to solve some numerical linear algebra problems;
3. use python to load, process, learn and interpret various data;
4. apply various factorization techniques to real world applications such as image processing, page rank, literature classification, etc;
5. write python code to implement iterative algorithms for solving large sparse systems, such as GMRES, CG, etc;
6. use Jypiter-note book to write technical reports and manuscripts, explaining numerical linear algebra topics in a way that combines theory and coding;
7. clearly communicate numerical linear algebra topics and their applications through oral presentations.
